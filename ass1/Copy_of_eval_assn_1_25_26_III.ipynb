```python
n_trials = 5

t_map = 0
d_map = 0
t_params = 0
t_train = 0
a_train = 0
a_test = 0
```

```python
for t in range( n_trials ):
  # Map the train points
  tic = time.perf_counter()
  X_map_trn = my_map( X_trn )
  toc = time.perf_counter()
  t_map += toc - tic
  d_map += X_map_trn.shape[ 0 ]

  # Get preferred hyperparameters
  tic = time.perf_counter()
  params = my_params( X_map_trn, X_trn, y_trn )
  toc = time.perf_counter()
  t_params += toc - tic

  # Create a new classifier object and set user-supplied hyperparameters
  clf = LinearSVC()
  clf.set_params( **params )

  # Train using the mapped features and find out train accuracy
  tic = time.perf_counter()
  clf.fit( X_map_trn, y_trn )
  toc = time.perf_counter()
  t_train += toc - tic
  a_train += clf.score( X_map_trn, y_trn )

  # Map the test points and find out test accuracy
  tic = time.perf_counter()
  X_map_tst = my_map( X_tst )
  toc = time.perf_counter()
  t_map += toc - tic
  a_test += clf.score( X_map_tst, y_tst )
```

```python
t_map /= n_trials
d_map /= n_trials
t_params /= n_trials
t_train /= n_trials
a_train /= n_trials
a_test /= n_trials

print( f"{t_map},{d_map},{t_params},{t_train},{a_train},{a_test}" )
```
